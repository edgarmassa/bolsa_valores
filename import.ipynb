```python
import yfinance as yf
import pandas as pd
import investpy as inv

import datetime as dte
from datetime import date
```

```python
# STOCKS = AÇÕES
br = inv.stocks.get_stocks(country='brazil')
br
```

```python
#br['symbol']

carteira = []
for a in br['symbol']:
    if len(a) <= 5:
        carteira.append(a + '.SA')
```

```python
#carteira
#dt_5

dt = date.today() 
dt_5 = dte.date( dt.year - 5, dt.month, dt.day )
```

```python
dt = yf.download(carteira, start=dt_5, end=dt)['Adj Close']
```

```python
df = pd.DataFrame(dt)
df
```
